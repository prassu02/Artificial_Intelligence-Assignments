```python
# import libraries
import pandas as pd
```

```python
# Load CSV
df = pd.read_csv('/content/employee.csv')
df.head()
```

## 1. Greet Employee by First Name and Gender

```python
def greet_employee(eid):
    emp = df[df['EID'] == eid]
    if emp.empty:
        return "Employee not found."

    name = emp.iloc[0]['Employee Name']
    gender = emp.iloc[0]['Gender']
    first_name = name.split()[0]

    if gender.upper() == 'M':
        return f"Hello, Mr {first_name}"
    elif gender.upper() == 'F':
        return f"Hello, Ms {first_name}"
    else:
        return f"Hello, {first_name}"

# Example
greet_employee("E123D")
```

## 2. Average Salary by Department

```python
def average_salary_by_department():
    avg_salary = df.groupby('Department')['Salary'].mean()
    return avg_salary

average_salary_by_department()
```

## 3. Get Employee Info by EID

```python
def get_employee_info(eid):
    emp = df[df['EID'] == eid]
    if emp.empty:
        return "Employee not found."

    name = emp.iloc[0]['Employee Name']
    doj = emp.iloc[0]['Date of Joining']
    dept = emp.iloc[0]['Department']

    return f"{name} | {doj} | {dept}"

# Example
get_employee_info("E123D")
```

## 4. Sort Employees from Oldest to Newest

```python
def sort_by_joining_date():
    df['Date of Joining'] = pd.to_datetime(df['Date of Joining'], dayfirst=True)
    sorted_df = df.sort_values('Date of Joining')
    return sorted_df[['EID', 'Employee Name', 'Date of Joining', 'Department']]

sort_by_joining_date().head()
```
